```python
import matplotlib.pyplot as plt
import numpy as np

# 这两行代码解决 plt 中文显示的问题
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 输入统计数据
waters = ('4', '16', '128', '256', '1024')
high = [116085,97755.4,109256,109256,116085]
low = [13860.8,14510.56,6705.24,6339.08,6170.605]
three = [8844.54,6030.36,5399.27,5202.67,4836.85]




bar_width = 0.3  # 条形宽度
index_male = np.arange(len(waters))
index_female = index_male + bar_width
index_three = index_female+bar_width

# 使用两次 bar 函数画出两组条形图
plt.bar(index_male, height=high, width=bar_width, color='mediumaquamarine', label='丢包率百分之0')
plt.bar(index_female, height=low, width=bar_width, color='salmon', label='丢包率百分之5')
plt.bar(index_three, height=three, width=bar_width, color='blue', label='丢包率百分之十')

plt.legend()  # 显示图例
plt.xticks(index_male + bar_width/2, waters)  # 让横坐标轴刻度显示 waters 里的饮用水， index_male + bar_width/2 为横坐标轴刻度的位置
plt.ylabel('吞吐率：byte/s')  # 纵坐标轴标题
plt.title('窗口大小-吞吐率比对图')  # 图形标题

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 这两行代码解决 plt 中文显示的问题
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 输入统计数据
waters = ('4', '16', '128', '256', '1024')
high = [13860,6449,6339,6170,5915]
low = [3783,3722,3670,3627,3558]
three = [2180,2162,2150,2137,2132]

bar_width = 0.3  # 条形宽度
index_male = np.arange(len(waters))
index_female = index_male + bar_width
index_three = index_female+bar_width

# 使用两次 bar 函数画出两组条形图
plt.bar(index_male, height=high, width=bar_width, color='mediumaquamarine', label='延迟20ms')
plt.bar(index_female, height=low, width=bar_width, color='salmon', label='延迟50ms')
plt.bar(index_three, height=three, width=bar_width, color='blue', label='延迟100ms')

plt.legend()  # 显示图例
plt.xticks(index_male + bar_width/2, waters)  # 让横坐标轴刻度显示 waters 里的饮用水， index_male + bar_width/2 为横坐标轴刻度的位置
plt.ylabel('吞吐率：byte/s')  # 纵坐标轴标题
plt.title('router时延-吞吐率比对图')  # 图形标题

plt.show()
```
